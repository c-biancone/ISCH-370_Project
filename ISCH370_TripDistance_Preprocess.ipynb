```python
#import trips by distance data
import pandas as pd
trips_distance=pd.read_csv('/content/drive/My Drive/data_sets/Trips_by_Distance.csv')
trips_distance
```

```python
#check for null values
trips_distance.isnull().sum()
```

```python
trips_distance1=trips_distance
```

```python
len(trips_distance1["County Name"].unique())
```

```python
trips_distance1.dropna(inplace=True)
trips_distance1
```

```python
#take out unimportant columns
dropped_columns=['State FIPS', 'Row ID', 'Level', 'Week', 'Month']
trips_distance1.drop(dropped_columns, axis=1, inplace=True)
trips_distance1
```

```python
most_numberoftrips=trips_distance1.sort_values(by=['Number of Trips'], ascending=False)
most_numberoftrips
#sorted the data by most number of trips and it can be seen that CA has the most as a state
```

```python
least_numberoftrips=trips_distance1.sort_values(by=['Number of Trips'], ascending=True)
least_numberoftrips
```

```python
average = trips_distance.copy()
```

```python
average.sort_values(by='County FIPS', inplace = True, ascending = True)
average.set_index('County FIPS')
print(average['County FIPS'].unique())
# correct range of counties
print(len(average['County FIPS'].unique()))
# missing a county
```

```python
# want to average all of the rows for a single county FIPS
average = average.groupby(level=0)
average = average.agg('mean')
average
```

```python
#look at just NY travel data
ny_trips=trips_distance1.loc[trips_distance1['State Postal Code']=='NY']
ny_trips
```

```python
#looking at only monroe county
monroe_ny_trips=ny_trips.loc[ny_trips['County Name']=='Monroe County']
monroe_ny_trips
```

```python
most_recent=monroe_ny_trips.sort_values(by=['Date'], ascending=True)
most_recent
```

```python
most_recent.plot.line(y=['Number of Trips >=500'])
```

```python
#looking at the data of those that stay at home more than leave
stay_at_home=trips_distance1.query("`Population Staying at Home` > `Population Not Staying at Home`")
stay_at_home
```

```python
#histogram of the number of trips
import matplotlib.pyplot as plt
stay_at_home.boxplot(column='Number of Trips')
```

```python
#bar plot of the number of trips by each state
import matplotlib.pyplot as plt
plt.bar(stay_at_home['State Postal Code'], stay_at_home['Number of Trips'], color="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
ny_trips_groubpy_county_monroe_2019 = ny_trips[ny_trips["Date"].str.contains("2019")].groupby("County Name").get_group("Monroe County").sort_values(by="Date")
plt.plot(ny_trips_groubpy_county_monroe_2019["Date"], ny_trips_groubpy_county_monroe_2019["Number of Trips"], label="2019")
plt.xlabel('Number of Trips', fontsize=12)
plt.ylabel('Date', fontsize=12)
#plt.legend(loc='upper left')
plt.title(f'Date vs. Number of Trips\nMonroe County, NY 2019')
plt.rcParams['figure.figsize'] = (20, 5)
plt.xticks(rotation = 90)
plt.xticks(ticks=ny_trips_groubpy_county_monroe_2019["Date"][::5])
plt.ylim([2e6,5e6])
plt.show()
```
